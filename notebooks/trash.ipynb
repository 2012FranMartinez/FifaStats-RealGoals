```python
import pandas as pd
from thefuzz import process, fuzz

# Crear dos DataFrames de ejemplo
df1 = pd.DataFrame({
    'nombre': ['Juan Perez', 'Maria Garcia', 'Pedro Alvarez'],
    'valor': [10, 20, 30]
})

df2 = pd.DataFrame({
    'nombre': ['Juan P.', 'Maria G.', 'Alvarez, Pedro'],
    'valor': [100, 200, 300]
})

# Función para hacer el fuzzy merge
def fuzzy_merge(df1, df2, key1, key2, threshold=85):
    matches = []
    
    for name in df1[key1]:
        match = process.extractOne(name, df2[key2], scorer=fuzz.token_sort_ratio)
        if match and match[1] >= threshold:
            matches.append(match[0])
        else:
            matches.append(None)
    
    df1['best_match'] = matches
    return df1

# Llamar a la función para realizar el merge
result = fuzzy_merge(df1, df2, 'nombre', 'nombre')

# Mostrar el resultado
print(result)
```

```python
# Llamada a la función
df_fifa_real = fuzzy_merge(df_fifa_esp, df_real, 'short_name', 'NOMBRE')
df_fifa_real
```
