```python
import pandas as pd
```

```python
# DATOS FIFA COMPLETOS
df_fifa_22_total = pd.read_csv("./data/players_22.csv")
```

```python
# Columnas a quedarnos
columnas_conservar = ["sofifa_id","short_name","long_name","club_position","overall","potential"
                      ,"pace","shooting","attacking_finishing","power_shot_power","power_long_shots","attacking_heading_accuracy","attacking_volleys",
                      "mentality_aggression","skill_fk_accuracy","mentality_penalties","dribbling","physic",
                      "attacking_short_passing","skill_ball_control","movement_acceleration",
                      "movement_sprint_speed","movement_agility","movement_reactions","movement_balance","power_jumping",
                      "power_strength","mentality_positioning","mentality_vision",
                      "preferred_foot","weak_foot","international_reputation",
                      "value_eur","age","height_cm","weight_kg","club_team_id","club_name","league_name","league_level","club_jersey_number",
                      "nationality_id","nationality_name",]
```

```python
# DATOS FIFA FILTRADOS
df_fifa_filtrado = df_fifa_22_total.filter(items=columnas_conservar)
```

```python
# DATOS FIF SOLO ESP
df_fifa_esp = df_fifa_filtrado[df_fifa_filtrado["league_name"] == "Spain Primera Division"]
df_fifa_esp
```

```python
# SOLO ID Y NOMBRE
cols_solo_nombres = ["sofifa_id","short_name","long_name"]
df_fifa_solo_nombres = df_fifa_esp.filter(items=cols_solo_nombres)
```

```python
# DATOS REALES
df_reales = pd.read_csv("DatosReales_12_10_2024.csv")
```

```python
# Eliminar columna unnamed
df_reales.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
# GOLEADORES REALES (y portería invatida)
df_reales_goleadores = df_reales[df_reales["GOLES_PORTERIA"]>0]
```
